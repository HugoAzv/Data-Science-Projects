# Given Canada's Adjusted net national income per capita, we are going to predict the income in the next years

```python
# importando as bibliotecas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import linear_model

sns.set()
plt.style.use("dark_background")
```

```python
df = pd.read_csv('canada_per_capita_income.csv')
df.head(5)
```

```python
years = df.iloc[:,0].values
pc_income = df.iloc[:,1].values
```

```python
years
```

```python
pc_income
```

# Análise Exploratória

```python
plt.figure(figsize=(20,9))
plt.scatter(years, pc_income, color='orange', marker='+');
plt.xlabel('Year')
plt.ylabel('Per Capita Income (US$)')
plt.grid(False)
```

# Criando o Modelo

```python
reg = linear_model.LinearRegression() # Creating the model
```

```python
x = years.reshape(-1, 1)
y = pc_income.reshape(-1, 1)

reg.fit( x, y ) # fitting the data to the model
```

```python
m = reg.coef_ # m
b = reg.intercept_
m, b
```

```python
to_predict = np.array([2020])
reg.predict(to_predict.reshape(-1,1))
```

```python
y = m*2020 + b
y
```

## Visualizando a previsão

```python
plt.figure(figsize=(20,9))
plt.scatter( x, y, color = 'orange' );
plt.xlabel('Year')
plt.ylabel('Per Capita Income (US$)')
plt.plot( x , reg.predict(x) , color='red' );
```
